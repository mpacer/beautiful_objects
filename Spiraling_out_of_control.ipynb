```python
from allofplos import starterdir, Corpus

from article_utils import style_punct, body_punct, get_punct
corpus = Corpus(starterdir)
```

```python
from spiral import ArchemedianSpiral, svgStyledSpiral, svgAnimateSpiral
```

```python
spiral = ArchemedianSpiral(a=20, b=2, num_cycles=10, angle=20, font_size=.75)
```

```python
def pretty_corpus(corpus):
    svg = svgAnimateSpiral(content=body_punct(corpus.random_article), 
                           spiral=spiral)        
    return svg.to_html()

svg = get_ipython().display_formatter.formatters['image/svg+xml']
svg.for_type(Corpus, pretty_corpus)
```

```python
corpus
```

```python
def animated_corpus(corpus):

    rep = []
    for idx, article in enumerate(corpus.random_sample(3)):
        svg = svgAnimateSpiral(content=style_punct(article), 
                               spiral=spiral,
                               delay=idx*2,
                               start_anim=True,
#                                rotate_anim=True
                              )
        rep += svg.to_html()
        
    return "".join(rep)

html = get_ipython().display_formatter.formatters['text/html']
html.for_type(Corpus, animated_corpus)
```
